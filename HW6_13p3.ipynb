# Homework 6, problem 13.3

Consider the polynomial

$$\begin{align}
p(x)&=
(x-2)^9\\
&=x^9-18x^8+144x^7-672x^6+2016x^5\\
&\quad\quad-4032x^4+5376x^3-4608x^2+2304x-512
\end{align}$$

```python
# import the required libraries
import numpy as np
import matplotlib.pyplot as plt

# set image parameters
plt.rcParams['figure.figsize'] = [16,8]
plt.rcParams['figure.dpi'] = 80

plt.rcParams['font.size'] = 25
plt.rcParams['legend.fontsize'] = 'large'
plt.rcParams['figure.titlesize'] = 'large'
```

### (a) Plot $p(x)$ for $x= 1.920, 1.921,1.922,\dots,2.080$, evaluating $p$ via its coefficients. 

```python
x = np.arange(1.92,2.081,0.001)
def p_coef(x):
    return(x**9-18*x**8+144*x**7-672*x**6+2016*x**5-4032*x**4+5376*x**3-4608*x**2+2304*x-512)
plt.plot(x,p_coef(x),'o')
plt.title('$p(x)$ using coefficients')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

### (b) Produce the same plot again, now evaluating $p$ via the factored form

```python
x = np.arange(1.92,2.081,0.001)
def p(x):
    return((x-2)**9)
plt.plot(x,p(x),'o')
plt.title('$p(x)$ using factored form')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

### Plot the error between the two evaluatings of $p$

```python
plt.plot(x,np.abs(p_coef(x)-p(x)),'o')
plt.title('Error between (a) and (b)')
plt.xlabel('$x$')
plt.ylabel('Error, (log scale)')
plt.yscale('log')
plt.show()
```
